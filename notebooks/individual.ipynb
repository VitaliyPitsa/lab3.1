<h3>Задача 1:</h3>
Один автомобиль, двигаясь равномерно со скоростью x м/с, в течение 10 с прошел такой же путь, какой второй автомобиль прошел за 15 с. Определить скорость второго автомобиля.

```python
from matplotlib import pylab as plt
%matplotlib inline
```

```python
v1 = float(input("Введите скорость  первого автомобиля: "))
z = v1 * 10
v2 = (v1 * 10)/15
t = z / v1
print(f"расстояние которое прошли автомобили: {z}")
print(f"скорость второго автомобиля: {v2}")
```

```python
x = [i for i in range(int(t) + 5)]
y1 = [(v1*i) for i in range(int(t) + 5)]
y2 = [(v2*i) for i in range(int(t) + 5)]
plt.plot(x, y1)
plt.plot(x, y2)
```

<h3>задача 2:</h3> 
скорость самолёта вычисляется по формуле: 
$$
V(t)=V_{0}t^{3}-\frac{gt^2}{2}+\sin(\alpha)t
$$


найти расстояние которое прошёл самолёт за t=5 если 

```python
import numpy as np
from scipy.integrate import quad


v0 = float(input("Введите начальную скорость самолёта: "))
a = np.radians(30) 
g = 9.81 
t = 5 
def V(t):
    return v0*t**3 - g*t**2/2 + np.sin(a)*t

def x_integrand(t):
    return V(t)

x = quad(x_integrand, 0, t)[0]
print(f"расстояние которое прошёл самолёт: {x}")
time = np.arange(0, 5, 0.01)
speed = [V(t) for t in time]
plt.plot(time, speed)
plt.xlabel('Time, s')
plt.ylabel('Speed, m/s')
plt.show()

```
